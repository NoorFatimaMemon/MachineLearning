# **Importing All Necessary Libraries**

```python
#Importing all dependencies and Seaborn itself
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Data Loading**

```python
df = pd.read_csv('HousingData.csv')
df.head()
```

```python
# Check data shape
df.shape
```

```python
# Check null values
df.isnull().sum()
```

# **Pair Plotting to show the relationship**

```python
# Create a pair plot
sns.pairplot(df)
plt.show()
```

# **KDE Plotting to know which column needs StandardScaler and MinMaxScaler**

```python
features = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
```

```python
for feature in features:
    sns.kdeplot(df[feature], shade=True)
    plt.title(f'KDE Plot for {feature}')
    plt.xlabel(f'{feature}')
    plt.ylabel('Density')
    plt.show()
```

# **Imputation**

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit_transform(df)
```

# **Using MinMaxScaler for all columns except RM (RM is normally distributed)**

```python
from sklearn.preprocessing import MinMaxScaler
columns_to_scale = df.columns[df.columns != 'RM']
scaler = MinMaxScaler()

# Fit and transform the columns to be scaled
df_scaled = df.copy()
df_scaled[columns_to_scale] = scaler.fit_transform(df_scaled[columns_to_scale])
```

# **Plotting KDE for all columns except 'RM'**

```python
for column in columns_to_scale:
    sns.kdeplot(df_scaled[column], label=column, shade=True)
    plt.title(f'KDE Plot of Scaled {column}')
    plt.xlabel('Scaled Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

# **Using StandardScaler for column RM (b/c RM is normally distributed)**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

# Fit and transform the 'RM' column
df_scaled['RM'] = scaler.fit_transform(df[['RM']])
```

# **Plotting KDE for column 'RM'**

```python
sns.kdeplot(df_scaled['RM'], label='RM', shade=True)
plt.title('KDE Plot of Scaled RM Column')
plt.xlabel('Scaled Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
